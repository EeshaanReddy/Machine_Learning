```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score
from tkinter import *
```

```python
nasa_data = pd.read_csv('Downloads/nasa_dataset.csv')
```

```python
nasa_data
```

```python
nasa_data.head()
```

```python
nasa_data.info()
```

```python
nasa_data["label"] = nasa_data["koi_pdisposition"].apply(
    lambda x: 1 if x == "CANDIDATE" else 0
)
nasa_data["label_2"] = nasa_data["koi_disposition"].apply(
    lambda x: 2 if x == "CONFIRMED" else 1 if x == "CANDIDATE" else 0
)
```

```python
nasa_data.drop(columns=['kepler_name','rowid','kepid','kepoi_name','koi_teq_err1','koi_teq_err2','koi_disposition','koi_pdisposition','koi_tce_delivname'], inplace=True)
```

```python
nasa_data.dropna(inplace=True)
```

```python
nasa_data.head()
```

```python
def clean_dataset(df):
    assert isinstance(nasa_data, pd.DataFrame),nasa_data.dropna(inplace=True)
    indices = ~nasa_data.isin([np.nan, np.inf, -np.inf]).any(1)
    return nasa_data[indices].astype(np.float64)

clean_dataset(nasa_data)
```

```python
x_data = nasa_data.drop(columns=['label','label_2'])
y_label_1 = nasa_data.label
y_label_2 = nasa_data.label_2
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_label_1, random_state=0, test_size=.40)
```

```python
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

```python
bagging_model = BaggingClassifier(base_estimator=RandomForestClassifier(n_estimators=100,criterion='gini'),n_estimators=10, random_state=0)
bagging_model.fit(X_train,y_train)
y_pred_bagging = bagging_model.predict(X_test)
```

```python
print('Accuracy: ' + str(metrics.accuracy_score(y_test, y_pred_bagging)))
print('\nConfusion Matrix:')
print(' TN,  FP, FN, TP')
clf_b = confusion_matrix(y_test, y_pred_bagging).ravel()
print(clf_b)
```

```python
randf = RandomForestClassifier(n_estimators=100, criterion='gini')
randf.fit(X_train, y_train)
y_pred_forest = randf.predict(X_test)
```

```python
print('Accuracy: ' + str(metrics.accuracy_score(y_test, y_pred_forest)))
print('\nConfusion Matrix:')
print(' TN,  FP, FN, TP')
clf_rf = confusion_matrix(y_test, y_pred_forest).ravel()
print(clf_rf)
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)
```

```python
print('Accuracy: ' + str(metrics.accuracy_score(y_test, y_pred_tree)))
print('\nConfusion Matrix:')
print(' TN,  FP, FN, TP')
clf_dt = confusion_matrix(y_test, y_pred_tree).ravel()
print(clf_dt)
```
